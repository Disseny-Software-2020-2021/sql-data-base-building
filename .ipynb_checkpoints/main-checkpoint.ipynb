```python
dic = {}
for i in range(len(old_HDD)):
    dic[old_HDD.iloc[i].title]=old_HDD.iloc[i].category_id   

def hola(x):
    try: return dic[x]
    except: return None
    
film.category_id = film.category_id.apply(hola)
```
